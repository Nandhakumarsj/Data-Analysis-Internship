<h2> Importing Modules </h2>

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns

%matplotlib inline
```

---
## Loading Iris Dataset

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
print(df.head(5))
y = iris.target
print(y)
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.25)
df.info()
```

---
## Visualizing Dataset

```python
df['iris'] = y
sns.pairplot(df, markers='o', height=2.5, hue='iris') # Based on Iris separated by hue
df = df.drop(columns=['iris'])
sns.pairplot(df, markers='o', height=2.5)
plt.show()
```

---
## Creating And Training Decision Tree Model

```python
dtree = DecisionTreeClassifier()
print('Decision Tree Created')
dtree.fit(X_train,y_train)
```

---
## Visualizing The Decision Tree Model

```python
classes = ['Setosa', 'Versicolor', 'Virginica']

plt.figure(figsize=(6,5), dpi=300)
plot_tree(dtree, feature_names=iris.feature_names, class_names=classes, filled=True, rounded=True)
plt.show()
```

---
## Testing and Evaluating Decision Tree Model

```python
dtree_y_pred = dtree.predict(X_test)
print("Classification Report for Decision Tree Model using Iris Dataset:")
print(classification_report(y_test,dtree_y_pred))
print("Scores for Decision Tree Model:",dtree.score(X_test, y_test))
```
